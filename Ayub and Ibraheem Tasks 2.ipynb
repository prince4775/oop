## Star Conversion to Delta

```python
class Star:
    def __init__(self, Ra=0.0, Rb=0.0, Rc=0.0, R1=0.0, R2=0.0, R3=0.0):
        self.Ra = Ra
        self.Rb = Rb
        self.Rc = Rc
        self.R1 = R1
        self.R2 = R2
        self.R3 = R3

    def set_values(self):
        self.R1 = float(input("Enter the value of R1: "))
        self.R2 = float(input("Enter the value of R2: "))
        self.R3 = float(input("Enter the value of R3: "))

    def calculate_delta(self):
        self.Ra = (self.R2 * self.R3)+(self.R1 * self.R2)+(self.R1 * self.R3) / self.R1 
        self.Rb = (self.R2 * self.R3)+(self.R1 * self.R2)+(self.R1 * self.R3) / self.R2
        self.Rc = (self.R2 * self.R3)+(self.R1 * self.R2)+(self.R1 * self.R3) / self.R3

    def display(self):
        print(f"Star Network Resistances: R1={self.R1} ohms, R2={self.R2} ohms,
              R3={self.R3} ohms")
        print(f"Delta Network Resistances: Ra={self.Ra} ohms, Rb={self.Rb} ohms,
              Rc={self.Rc} ohms")


# Create objects and initialize them
star1 = Star()
star2 = Star(0.0, 0.0, 0.0, 1.0, 2.0, 3.0)

# Take input for the first object and calculate delta values
star1.set_values()
star1.calculate_delta()

# Display results for the first object
star1.display()

```

## 2D and 3D Class

```python
import math

class TwoD:
    def __init__(self, i, j):
        self.x = i
        self.y = j

    def set(self, X, Y):
        self.x = X
        self.y = Y

    def display(self):
        print(f"The coordinates you entered are ({self.x} ,{self.y})")

class ThreeD(TwoD):
    def __init__(self, i, j, k):
        super().__init__(i, j)
        self.z = k

    def set(self, X, Y, Z):
        super().set(X, Y)
        self.z = Z

    def display(self):
        print(f"The coordinates are ({self.x} ,{self.y} ,{self.z})")

    def cal(self, other):
        distance = math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2)
        return distance

# Create objects and initialize them
obj1 = TwoD(1.0, 2.0)
obj1.display()

obj2 = ThreeD(3.0, 4.0, 5.0)
obj2.set(6.0, 7.0, 8.0)
obj2.display()

obj3 = ThreeD(2.0, 4.0, 7.0)
obj3.display()

# Calculate distance between obj2 and obj3
distance = obj2.cal(obj3)
print(f"Distance between obj2 and obj3: {distance}")
```

## Hierarchy in a university


```python
class Student:
    def __init__(self, reg_number, name, gpa):
        self.reg_number = reg_number
        self.name = name
        self.gpa = gpa

    def set_info(self, reg_number, name, gpa):
        self.reg_number = reg_number
        self.name = name
        self.gpa = gpa

    def display(self):
        print(f"Registration Number: {self.reg_number}")
        print(f"Student Name: {self.name}")
        print(f"GPA: {self.gpa}")

class Undergraduate(Student):
    def __init__(self, reg_number, name, gpa, year):
        super().__init__(reg_number, name, gpa)
        self.year = year

    def set_info(self, reg_number, name, gpa, year):
        super().set_info(reg_number, name, gpa)
        self.year = year

    def display(self):
        super().display()
        print(f"Year: {self.year}")

class Graduate(Student):
    def __init__(self, reg_number, name, gpa, thesis_topic):
        super().__init__(reg_number, name, gpa)
        self.thesis_topic = thesis_topic

    def set_info(self, reg_number, name, gpa, thesis_topic):
        super().set_info(reg_number, name, gpa)
        self.thesis_topic = thesis_topic

    def print_info(self):
        super().display()
        print(f"Thesis Topic: {self.thesis_topic}")


# Declare object of type Undergraduate

student = Student("22jzele0470", 'Muhammad Ayub', 3.8)
student.display()

undergrad_student = Undergraduate("22jzele0480", 'Ibraheem khan', 3.9 , 2)
undergrad_student.display()

# Declare object of type Graduate
grad_student = Graduate("22jzele0466", 'Mashal Ahmad', 3, "Machine Learning")
grad_student.print_info()
```

## Publishing company 

```python
class Publication:
    def __init__(self):
        self.title = ""
        self.price = 0.0

    def getdata(self):
        self.title = input("Enter the title: ")
        self.price = float(input("Enter the price: "))

    def display(self):
        print(f"Title: {self.title}")
        print(f"Price: {self.price}")


class Book(Publication):
    def __init__(self):
        super().__init__()
        self.page_count = 0

    def getdata(self):
        super().getdata()
        self.page_count = int(input("Enter the page count: "))

    def display(self):
        super().display()
        print(f"Page Count: {self.page_count}")


class Tape(Publication):
    def __init__(self):
        super().__init__()
        self.playing_time = 0.0

    def getdata(self):
        super().getdata()
        self.playing_time = float(input("Enter the playing time in minutes: "))

    def display(self):
        super().display()
        print(f"Playing Time: {self.playing_time} minutes")

 # Test the Book class
book_instance = Book()
print("Enter information for the book:")
book_instance.getdata()
print("\nDisplaying book information:")
book_instance.display()

print("\n--------------------------\n")

# Test the Tape class
tape_instance = Tape()
print("Enter information for the tape:")
tape_instance.getdata()
print("\nDisplaying tape information:")
tape_instance.display()


```
