```python
# Define the grade points for each letter grade
grade_points = {
    'A': 4.0,
    'A-': 3.7,
    'B+': 3.3,
    'B': 3.0,
    'B-': 2.7,
    'C+': 2.3,
    'C': 2.0,
    'C-': 1.7,
    'D+': 1.3,
    'D': 1.0,
    'F': 0.0,
}

# Define the courses, letter grades, and credit hours for two semesters
semester1 = [
    {'course': 'calculas', 'grade': 'D', 'credits': 3},
    {'course': 'EM', 'grade': 'F', 'credits': 3},
    {'course': 'c_f', 'grade': 'F', 'credits': 2},
    {'course': 'ethics', 'grade': 'C', 'credits': 2},
    {'course': 'islamiyat', 'grade': 'B', 'credits': 2},
    {'course': 'pak_study', 'grade': 'F', 'credits': 2},
    {'course': 'c_f_lab', 'grade': 'B-', 'credits': 1},
    {'course': 'em_lab', 'grade': 'A-', 'credits': 1},
    {'course': 'eng_lab', 'grade': 'B+', 'credits': 1},
    # Add more courses and grades here
]

semester2 = [
    {'course': 'computer_Programing', 'grade': 'D', 'credits': 3},
    {'course': 'computer_Programing_Lab', 'grade': 'D', 'credits': 1},
    {'course': 'Basic_Electrical_Engineering', 'grade': 'B-', 'credits': 3},
    {'course': 'Basic_Electrical_Engineering_Lab', 'grade': 'B+', 'credits': 1},
    {'course': 'linear_Algebra', 'grade': 'A-', 'credits': 3},
    {'course': 'Engineering_Mechanics', 'grade': 'B+', 'credits': 3},
    {'course': 'Workshop_Technology', 'grade': 'C', 'credits': 1},
    {'course': 'Workshop_Technology_Lab', 'grade': 'B+', 'credits': 1},
    {'course': 'Composition_and_Comprehension', 'grade': 'B', 'credits': 3},
    # Add more courses and grades here
]

# Function to calculate the GPA for a semester
def calculate_semester_gpa(semester):
    total_credits = 0
    total_grade_points = 0

    for course in semester:
        grade = course['grade']
        credits = course['credits']

        total_credits += credits
        total_grade_points += grade_points.get(grade, 0) * credits

    if total_credits == 0:
        return 0.0

    gpa = total_grade_points / total_credits
    return gpa

# Calculate the GPAs for both semesters
gpa1 = calculate_semester_gpa(semester1)
gpa2 = calculate_semester_gpa(semester2)

print(f'GPA for Semester 1: {gpa1:.2f}')
print(f'GPA for Semester 2: {gpa2:.2f}')
Sum_gpa=gpa1+gpa2
Cgpa=Sum_gpa/2
print(f'My Cgoa is = : {Cgpa}')
```

```python
class GPA_Calculator:
    def __init__(self):
        self.course_data = []
    
    def add_course(self, course_name, credits, grade):
        self.course_data.append((course_name, credits, grade))
    
    def calculate_gpa(self):
        total_quality_points = 0
        total_credits = 0
        
        for course in self.course_data:
            credits = course[1]
            grade = course[2]
            quality_points = self.grade_to_quality_points(grade)
            
            total_quality_points += credits * quality_points
            total_credits += credits
        
        if total_credits == 0:
            return 0.0
        else:
            gpa = total_quality_points / total_credits
            return gpa
    
    def grade_to_quality_points(self, grade):
        grade_points = {
            "A": 4.0, "A-": 3.7,
            "B+": 3.3, "B": 3.0, "B-": 2.7,
            "C+": 2.3, "C": 2.0, "C-": 1.7,
            "D+": 1.3, "D": 1.0, "F": 0.0
        }
        return grade_points.get(grade, 0.0)

def main():
    calculator = GPA_Calculator()
    
    while True:
        course_name = input("Enter course name (or 'q' to quit):` ")
        course_name = course_name.upper()
        if course_name == 'Q':
            break
        
        credits = float(input("Enter credits: "))
        grade = input("Enter grade: ")
        grade=grade.upper()
        
        calculator.add_course(course_name, credits, grade)
    
    gpa = calculator.calculate_gpa()
    print(f"Your GPA is: {gpa:.2f}")

if __name__ == "__main__":
    main()
```
