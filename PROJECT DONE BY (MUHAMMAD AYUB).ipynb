# Normalize the data
At this step we will noramlize the data and then plot the both data orignal and normalized data, we will noramlize it with the log function which is in the numpy library In the NumPy library, the `log` function is used to calculate the natural logarithm (base e) of the elements in an array. It is part of the mathematical functions available in NumPy and is particularly useful when dealing with data that follows exponential or logarithmic patterns. Here's how you can use the `log` function. 

The `log` function allows you to transform data and perform operations that are based on logarithmic scales, which is common in various scientific and mathematical contexts. 

```python
# Add a small constant to ensure all values are positive
shifted_data = df + 1e-5  # You can adjust the constant as needed

# Apply a logarithmic transformation to the shifted data
Normalize_data = np.log(shifted_data)

# Plot the original and normalized data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df, bins=30, alpha=0.7)
plt.title('Original Right-Skewed Data')

plt.subplot(1, 2, 2)
plt.hist(Normalize_data, bins=30,  alpha=0.7)
plt.title('Normalized (Shifted and Log-transformed) Data')

plt.tight_layout()
plt.show()
```

# Splitting Data in train & test
At this step we will split the whole data into two parts (train and test) and select the features and the target , train the model with the 80% of data and test it with the 20% data after this we will train the model which will predict the target and also note that we will train the different models wnd at the last we will select the best accuracy model

```python
features=df.drop("price_range",axis=1)
target=df["price_range"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,target,train_size=0.8)
```

# Training of model by Decision Tree
A decision tree model is a supervised machine learning algorithm used for both classification and regression tasks. It represents decisions and their potential consequences in a tree-like structure, where each internal node corresponds to a feature or attribute, and each leaf node represents a class label or a numeric value. The tree is built by recursively splitting the dataset into subsets based on the most informative features, optimizing criteria like entropy or Gini impurity. Decision trees are interpretable and versatile, making them valuable for both understanding data relationships and making predictions in various domains, such as finance, healthcare, and recommendation systems.
we will train the model first with the decision tree but it has some drawbacks such as ( overfitting or underfitting ) which is not required in the model because it will not predict in the real world problems by overfitting and alo it will waste alot of time

```python
clf = DecisionTreeClassifier(random_state=42)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

```

# Checking of accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

# Selecting of best features
`chi2` (Chi-squared) is a statistical test used to measure the independence of two categorical variables. In the context of feature selection, it can be applied with `SelectKBest` to select the top K features(k  means number of features which will select) with the highest chi-squared statistics, indicating the strongest relationships between input features and the target variable. This helps improve machine learning model performance by focusing on the most informative features. 

```python

# Create a SelectKBest instance with chi-squared scoring
k_best = SelectKBest(score_func=chi2, k=8)  # Select the top 2 features

# Fit the feature selector on the training data and transform the data
X_train_selected = k_best.fit_transform(X_train, y_train)
X_test_selected = k_best.transform(X_test)


# Get the indices of the selected features
selected_feature_indices = k_best.get_support(indices=True)
selected_features = features.columns[selected_feature_indices]

print("Selected features:")
print(selected_features)
```

# Price prediction by Randomforestclassifier
Random Forest is an ensemble machine learning algorithm that combines multiple decision trees to improve predictive accuracy and reduce overfitting. It works by creating a collection of decision trees during training, each trained on a random subset of the data and using a random subset of features. During prediction, the individual tree predictions are combined (typically by averaging for regression and voting for classification) to produce a more robust and accurate result. Random Forest is highly versatile, handles both categorical and numerical data well, and is less prone to overfitting compared to single decision trees. It's widely used for tasks like classification, regression, and feature selection in data science and machine learning.In that problem we will use the randomforest for the prediction of price range of mobile phones which is our target 

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features[['battery_power', 'fc', 'int_memory', 'mobile_wt', 'px_height',
       'px_width', 'ram', 'talk_time']], target, test_size=0.2, random_state=42)

# Create a Random Forest classifier
clf = RandomForestClassifier(random_state=42)

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")


predicted_price_range = clf.predict([[1000,2.5,64,10,10,4,4,10]])

print(f"Predicted Price Range for the new mobile phone: {predicted_price_range[0]}")
```
