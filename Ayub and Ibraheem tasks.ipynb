## Series circuit

```python
class SeriesCircuit:
    def __init__(self):
        self.number_of_resistance = 0
        self.net_resistance = 0
        self.resistances = []

    def __init__(self, num_resistance, val_resistance):
        self.number_of_resistance = num_resistance
        self.net_resistance = 0
        self.resistances = val_resistance

    def get_val(self):
        self.number_of_resistance = int(input("Enter the number of resistances: "))
        self.resistances = [float(input(f"Enter the resistance {i + 1} in ohms: "))
                            for i in range(self.number_of_resistance)]

    def calc_net_res(self):
        self.net_resistance = sum(self.resistances)

    def display(self):
        print(f"Number of resistances: {self.number_of_resistance}")
        print(f"Resistance values: {self.resistances}")
        print(f"Net resistance: {self.net_resistance} ohms")


# Main function
'''obj1 = SeriesCircuit()
obj1.get_val()
obj1.calc_net_res()
obj1.display()'''

num_res = int(input("Enter the number of resistances for obj2: "))
val_res = [float(input(f"Enter the resistance {i + 1} in ohms: ")) for i in range(num_res)]
obj2 = SeriesCircuit(num_res, val_res)
obj2.calc_net_res()
obj2.display()
```

## Parallel circuit

```python
class parallel:
    def __init__(self):
        self.number_of_resistance = 0
        self.net_resistance = 0
        self.resistances = []

    def __init__(self, num_resistance, val_resistance):
        self.number_of_resistance = num_resistance
        self.net_resistance = 0
        self.resistances = val_resistance

    def get_val(self):
        self.number_of_resistance = int(input("Enter the number of resistances: "))
        self.resistances = [float(input(f"Enter the resistance {i + 1} in ohms: ")) 
                            for i in range(self.number_of_resistance)]

    def calculate_net_resistance(self):
        if self.number_of_resistance > 0:
            reciprocal_sum = sum(1 / resistance for resistance in self.resistances)
            self.net_resistance = 1 / reciprocal_sum
        else:
            self.net_resistance = 0

    def display(self):
        print(f"Number of resistances: {self.number_of_resistance}")
        print(f"Resistance values: {self.resistances}")
        print(f"Net resistance: {self.net_resistance} ohms")


# Main function
'''obj1 = parallelCircuit()
obj1.get_val()
obj1.calc_net_res()
obj1.display()'''

num_res = int(input("Enter the number of resistances for obj2: "))
val_res = [float(input(f"Enter the resistance {i + 1} in ohms: ")) for i in range(num_res)]
obj2 = parallel(num_res, val_res)
obj2.calculate_net_resistance()
obj2.display()
```

## Quadratic

```python
import math

class Quadratic:
    def __init__(self, a=0, b=0, c=0):
        self.a = a
        self.b = b
        self.c = c

    def set_values(self):
        self.a = float(input("Enter the value for a: "))
        self.b = float(input("Enter the value for b: "))
        self.c = float(input("Enter the value for c: "))

    def calculate_roots(self):
        discriminant = self.b**2 - 4 * self.a * self.c

        if discriminant > 0:
            root1 = (-self.b + math.sqrt(discriminant)) / (2 * self.a)
            root2 = (-self.b - math.sqrt(discriminant)) / (2 * self.a)
            print("Roots are real and different.")
            print("Root 1 =", root1)
            print("Root 2 =", root2)
        elif discriminant == 0:
            root = -self.b / (2 * self.a)
            print("Roots are real and the same.")
            print("Root =", root)
        else:
            real_part = -self.b / (2 * self.a)
            imaginary_part = math.sqrt(-discriminant) / (2 * self.a)
            print("Roots are complex and different.")
            print(f"Root 1 = {real_part} + {imaginary_part}i")
            print(f"Root 2 = {real_part} - {imaginary_part}i")

    def display(self):
        print(f"Quadratic Equation: {self.a}x^2 + {self.b}x + {self.c}")


obj1 = Quadratic()

obj1.set_values()
obj1.display()
obj1.calculate_roots()
```

## Delta conversion

```python
class Delta:
    def __init__(self, Ra=0, Rb=0, Rc=0, R1=0, R2=0, R3=0):
        self.Ra = Ra
        self.Rb = Rb
        self.Rc = Rc
        self.R1 = R1
        self.R2 = R2
        self.R3 = R3

    def set_values(self):
        self.Ra = float(input("Enter the value for Ra in ohms: "))
        self.Rb = float(input("Enter the value for Rb in ohms: "))
        self.Rc = float(input("Enter the value for Rc in ohms: "))

    def calculate_star_resistances(self):
        self.R1 = (self.Rb * self.Rc) / (self.Ra + self.Rb + self.Rc)
        self.R2 = (self.Ra * self.Rc) / (self.Ra + self.Rb + self.Rc)
        self.R3 = (self.Ra * self.Rb) / (self.Ra + self.Rb + self.Rc)

    def display(self):
        print("Delta Network Resistances:")
        print(f"Ra = {self.Ra} ohms")
        print(f"Rb = {self.Rb} ohms")
        print(f"Rc = {self.Rc} ohms")
        print("\nStar Network Resistances:")
        print(f"R1 = {self.R1} ohms")
        print(f"R2 = {self.R2} ohms")
        print(f"R3 = {self.R3} ohms")


# Main program
obj1 = Delta()
obj1.set_values()
obj1.calculate_star_resistances()
obj1.display()
```

## Complex

```python
import cmath  # Complex math library for phase calculation

class ComplexNumber:
    def __init__(self, real=0, imag=0, magnitude=0, phase=0):
        self.real = real
        self.imag = imag
        self.magnitude = magnitude
        self.phase = phase

    def set_values(self):
        self.real = float(input("Enter the real part: "))
        self.imag = float(input("Enter the imaginary part: "))

    def calculate_magnitude_phase(self):
        self.magnitude = abs(complex(self.real, self.imag))
        self.phase = cmath.phase(complex(self.real, self.imag))

    def display(self):
        print(f"Phasor Form: {self.magnitude}∠{self.phase}")

# Main program
obj1 = ComplexNumber()
obj1.set_values()
obj1.calculate_magnitude_phase()
obj1.display()
```

## Velocity of satellite

```python
class Satellite:
    def __init__(self, G=0, R=0, velocity=0):
        self.G = G
        self.R = R
        self.velocity = velocity

    def set_values(self):
        self.G = 6.673 * (10 ** -11)
        self.R = float(input("Enter the radius of the orbit for the satellite (in meters): "))

    def calculate_velocity(self, M_central):
        self.velocity = (self.G * M_central / self.R) ** 0.5

    def display(self):
        print(f"Velocity of the satellite: {self.velocity} m/s")

obj1 = Satellite()
obj1.set_values()

M_central = float(input("Enter the mass of the central body (in kg): "))
obj1.calculate_velocity(M_central)

obj1.display()
```

## Fraction:

```python
class Fraction:
    def __init__(self, a=0, b=0, c=0, d=0, r=0):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        self.r = r

    def set_values(self):
        self.a = int(input("Enter the numerator for the first fraction: "))
        self.b = int(input("Enter the denominator for the first fraction: "))
        self.c = int(input("Enter the numerator for the second fraction: "))
        self.d = int(input("Enter the denominator for the second fraction: "))

    def calculate_sum(self):
        self.r = (self.a * self.d + self.b * self.c) / (self.b * self.d)

    def display(self):
        print(f"The sum of {self.a}/{self.b} and {self.c}/{self.d} is: {self.r}")


while True:
    obj = Fraction()
    obj.set_values()
    obj.calculate_sum()
    obj.display()

    choice = input("Do you want to continue? (y/n): ")
    if choice.lower() != 'y':
        break

```

## TankProcess

```python
import math

class TankProcess:
    def __init__(self, initial_salt=100, initial_water=1000, brine_in_rate=10, brine_out_rate=10):
        self.initial_salt = initial_salt
        self.initial_water = initial_water
        self.brine_in_rate = brine_in_rate
        self.brine_out_rate = brine_out_rate
        self.current_time = 0

    def calculate_salt_amount(self, time):
        self.current_time = time
        salt_amount = 5000 - 4900 * math.exp(-0.01 * self.current_time)
        return salt_amount

    

# Example usage with user input
tank = TankProcess()

# Get user input for time interval
time_interval = int(input("Enter the time interval in minutes: "))

amount_at_t0 = tank.calculate_salt_amount(time_interval)

print(f"Amount of salt after {time_interval} minutes: {amount_at_t0} lb")
```

## Phasor

```python
import cmath

class Phasor:
    def __init__(self):
        self.peak = 0
        self.phase = 0

    def __init__(self, peak, phase):
        self.peak = peak
        self.phase = phase

    def set_val(self):
        self.peak = float(input("Enter the peak value: "))
        self.phase = float(input("Enter the phase angle in degrees: "))

    def __repr__(self):
        print(f"Phasor: {self.peak} ∠{self.phase}°")

    def add_phasor(phasor1, phasor2):
        a= phasor1.peak + phasor2.peak
        ''' The expression cmath.rect(1, phasor1.phase) is using the rect 
        function from the cmath module in Python. This function is used 
        to create a complex number in rectangular (Cartesian) form using 
        its magnitude and phase. '''
        b = cmath.phase(cmath.rect(1, phasor1.phase) + cmath.rect(1, phasor2.phase))
        return f" {a}∠{b}"

    def mul_phasor(phasor1, phasor2):
        c = phasor1.peak * phasor2.peak
        d = phasor1.phase + phasor2.phase
        return f" {c}∠{d}"
  # Main function

# Creating objects
#phs1 = Phasor()
#phs1.set_val()

peak_val1 = float(input("Enter the peak value for phs1: "))
phase_angle1 = float(input("Enter the phase angle in degrees for phs1: "))
phs1 = Phasor(peak_val1, phase_angle1)

peak_val = float(input("Enter the peak value for phs2: "))
phase_angle = float(input("Enter the phase angle in degrees for phs2: "))
phs2 = Phasor(peak_val, phase_angle)



# Displaying phasors
phs1
phs2

# Adding phasors

print("Sum of Phasor:")
print(Phasor.add_phasor(phs1,phs2))


# Multiplying phasors

print("mul of Phasor:")

print(Phasor.mul_phasor(phs1,phs2))

```
