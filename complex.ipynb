```python
import cmath

class Complex:
    def __init__(self):
        self.real = 0.0
        self.imag = 0.0

    def set(self):
        self.real = float(input("Enter the real part of the complex number: "))
        self.imag = float(input("Enter the imaginary part of the complex number: "))

    def display(self):
        print(f"   {self.real}   {self.imag}")

    def add(self, c1, c2):
        self.real = c1.real + c2.real
        self.imag = c1.imag + c2.imag

    def sub(self, c1, c2):
        self.real = c1.real - c2.real
        self.imag = c1.imag - c2.imag

    def mul(self, c2, c3):
        print("MULTIPLICATION")
        self.real = c2.real * c3.real - c2.imag * c3.imag
        self.imag = c2.real * c3.imag + c2.imag * c3.real

class TimeDomain(Complex):
    def __init__(self):
        super().__init__()
        self.peak_voltage = 0.0
        self.phase = 0.0

    def set(self):
        super().set()
        self.peak_voltage = float(input("Enter the peak voltage: "))
        self.phase = float(input("Enter the phase: "))

    def display(self):
        super().display()
        print(f"Peak Voltage: {self.peak_voltage}")
        print(f"Phase: {self.phase}")

    def con(self):
        # Convert time domain values to phasor form
        polar_form = cmath.polar(complex(self.real, self.imag))
        self.peak_voltage, self.phase = polar_form

    def add(self, t1, t2):
        super().add(t1, t2)
        self.con()

    def sub(self, t1, t2):
        super().sub(t1, t2)
        self.con()

    def mul(self, t1, t2):
        super().mul(t1, t2)
        self.con()

class RLC(Complex):
    # Define appropriate data members and member functions for RLC
    pass

# Example usage
time_obj = TimeDomain()
rlc_obj = RLC()

# Example usage
print("Enter values for TimeDomain object:")
time_obj.set()
time_obj.display()

print("\nEnter values for RLC object:")
rlc_obj.set()
rlc_obj.display()
```

```python
class Publication:
    def __init__(self):
        self.title = ""
        self.price = 0.0

    def getdata(self):
        self.title = input("Enter the title: ")
        self.price = float(input("Enter the price: "))

    def display(self):
        print(f"Title: {self.title}")
        print(f"Price: {self.price}")


class Book(Publication):
    def __init__(self):
        super().__init__()
        self.page_count = 0

    def getdata(self):
        super().getdata()
        self.page_count = int(input("Enter the page count: "))

    def display(self):
        super().display()
        print(f"Page Count: {self.page_count}")


class Tape(Publication):
    def __init__(self):
        super().__init__()
        self.playing_time = 0.0

    def getdata(self):
        super().getdata()
        self.playing_time = float(input("Enter the playing time in minutes: "))

    def display(self):
        super().display()
        print(f"Playing Time: {self.playing_time} minutes")

 # Test the Book class
book_instance = Book()
print("Enter information for the book:")
book_instance.getdata()
print("\nDisplaying book information:")
book_instance.display()

print("\n--------------------------\n")

# Test the Tape class
tape_instance = Tape()
print("Enter information for the tape:")
tape_instance.getdata()
print("\nDisplaying tape information:")
tape_instance.display()


```
